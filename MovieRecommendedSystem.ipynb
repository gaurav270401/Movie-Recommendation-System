```python
# As we are creating content based recommendation system so we have to see which columns are useful to recommend the movie based on content.
# List of columns :genres,id,keyword,title,overview,cast,crew.
movies = movies[['movie_id','genres','keywords','title','overview','cast','crew']]
```

```python
#check missing and duplicates in above data frame.If missing value is there and not too much then remove that movies from data frame and same with duplicacy.
#Learn about data cleaning in machine learning in deep-while revision.
#check missing values:
movies.isnull().sum()
```

```python
#drop missing values
movies.dropna(inplace=True)
```

```python
#check duplicacy
movies.duplicated().sum()
```

```python
#remove duplicacy
movies = movies.drop_duplicates()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#To simplify the things for recommendation we reduce above dataframe to new dataframe which includes only 3 columns like:id,title and tags.In tags we just combine all the remaining attributes of above dataframe.
#1st step is to convert all the attributes which is in other weird form like here string oflist  to the form of list which will easy to merge.Like in geners we just required directly name of geners so remove all other things.
import ast
def remove(obj):
    l= []
    for i in ast.literal_eval(obj):
       l.append(i['name']) #The literal_eval safely evaluate an expression node or a string containing a Python literal or container display. The string or node (a file) provided may only consist of the following Python literal structures: strings, bytes, numbers, tuples, lists, dicts, sets, booleans, and None.
    return l               #ex.It converts stringoflist form into list.'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]' to [{'id': 28, 'name': 'Action'},{'id': 12, 'name': 'Adventure'},{'id': 14, 'name': 'Fantasy'},{'id': 878, 'name': 'Science Fiction'}]
```

```python
#stringoflist to list for geners
movies['genres'] = movies['genres'].apply(remove)
```

```python
#stringoflist to list for keywords
movies['keywords']= movies['keywords'].apply(remove)
```

```python
import ast
def remove1(obj):
    l= []
    cn=0
    for i in ast.literal_eval(obj):
        if cn>=3:
            break
        else:
            cn+=1
            l.append(i['name'])
    return l 
```

```python
#As most of the times we recommend the movie based on lead cast hence we considered only starting 3 actors of cast.
##stringoflist to list for cast
movies['cast']=movies['cast'].apply(remove1)
```

```python
movies.head(1).cast
```

```python
movies.head()['crew']
```

```python
import ast
def remove2(obj):
    l= []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name']) 
            break
    return l 
```

```python
#similiarly we mostly known about the director of the movie and on that basis only we recommend movies so thats why we just keep director in list of crew as above function.
movies['crew']=movies['crew'].apply(remove2)
```

```python
#to combine all other with overview to make it one we convert it into list
movies['overview']=movies['overview'].apply(lambda x:x.split()) #Python Lambda Functions are anonymous function means that the function is without a name.
```

```python
movies.head()
```

```python
#Now to avoid confusion between two people which have same first name or same last name we just combine full name into one word like'sam mendes' to 'sammendes' in cast and crew.
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
#similiarly genre of movies
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head(1)
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies['tag'].head(1)
```

```python
new_movies = movies[['movie_id','title','tag']]
```

```python
new_movies
```

```python
new_movies['tag']=new_movies['tag'].apply(lambda x:" ".join(x))
```

```python
new_movies['tag']=new_movies['tag'].apply(lambda x:x.capitalize())
```

```python
new_movies.head()
```

```python
import nltk  #this library used for steaming technique
```

```python
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer() 
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies['tag']=new_movies['tag'].apply(stem)
```

```python
#As we done with data preprocessing and we create tags for every movie now we have to recommend movies based on the similarity in the tags.
#for that we are making website.For finding similarity between tags we are using vectorization(text vectorization(t.v.)) concept.We use 'bags of words' technique in that text vectorization.
# Bags of words: It means we have to collect or concatenate all tags and find most common words suppose 1st 2000 or 5000 common words we have to find and then just check count of that words in tags of every movie by excluding stop words like 'of on to from' etc.every movie with that count is act as a vector and w.r.t that  we recommend.
# we can do these above process manually but we have inbult libraries
from sklearn.feature_extraction.text import CountVectorizer # countervectorizer is an inbuilt class which is used here for bag of words technique.
cv = CountVectorizer(max_features=5000,stop_words='english') #This class have no of entities but we needed only no of words and stop words from that no. of entities.
```

```python
cv.fit_transform(new_movies['tag']) #fit transform is responsible for creating matrix outlet for BoW.
```

```python
vector = cv.fit_transform(new_movies['tag']).toarray()#In layman's terms, fit_transform means to do some calculation and then do transformation (say calculating the means of columns from some data and then replacing the missing values).Here It gives us a sparse matrix of type class numpy and toarray() converts it into numpy array.as we require numpy array(vector) for further process.
#If most of the elements of the matrix have 0 value, then it is called a sparse matrix.Python’s SciPy gives tools for creating sparse matrices using multiple data structures, as well as tools for converting a dense matrix to a sparse matrix. 
```

```python
vector[0]
```

```python
cv.get_feature_names() #get_feature_names() gives finded common words names.
#Also we use stemming technique to make similiar type of words like love,loving,loved into love i.e it replace all the words with love as it have same meaning just in different tense.
```

```python
#Now we have to find distace between 2 vectors i.e 2 movies to get similarities between movies.That distance will be the cosine distance i.e angle between vectors not eucladian distance (point to point distance).
#whenever you work with high dimensional data as here 5000(movies) dimensional data always calculate cosine distance as euclarian distance is not reliable.
#for this we use inbuilt function cosine similarity from sklearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
#now process of recommendation be like 1.find the index of the given movie 2. from that we get similarity array for that index(movie).
#3.To maintain the index with the similarity we will make a list of index with their similarity.4.Then we will sort that array in descending order w.r.t similarity.5.after that we will recommend whatever no of movies we want. 
#ex.If we want to recommend based on the avatar movie so we find its index which is 0.From that we jump to similarity matrix where we get a vector which gives similarity of all movies w.r.t avatar. After that to get most similar movies we have to sort that similarity matrix in decreasing order.
#But after sort its very difficult to get index of that most similar movies as after sorting it changes its position in vector.So To maintain the index with the similarity we will make a list of Tuple(index with their similarity).Here we use enumerate function for that.
def recommend(movie):
    movie_index=new_movies[new_movies['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
     
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
    
    
    
    
```

```python
recommend('Avatar')
```
